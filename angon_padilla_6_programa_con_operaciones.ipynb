```python
class EstacionMeteorologica:

    @staticmethod
    def convertir_a_celsius(temperaturas_fahrenheit):
        return [(temp - 32) * 5/9 for temp in temperaturas_fahrenheit]

    @staticmethod
    def mostrar_temperaturas(temperaturas_fahrenheit):
        temperaturas_celsius = EstacionMeteorologica.convertir_a_celsius(temperaturas_fahrenheit)
        print("Temperaturas en Fahrenheit:", temperaturas_fahrenheit)
        print("Temperaturas en Celsius:", temperaturas_celsius)


temperaturas_fahrenheit = [80, 72, 95, 82, 79, 72, 85]

EstacionMeteorologica.mostrar_temperaturas(temperaturas_fahrenheit)
```
