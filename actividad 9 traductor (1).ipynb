Agregamos 10 palabras al diccionario para usarlas despues

```python
dic = {"hola": "hello", "adios": "goodbye", "gracias": "thank you", "por favor": "please", "si": "yes", "no": "no", "agua": "water", "comida": "food", "sol": "sun", "luna": "moon"}
```

Creamos una funcion para traducir las palabras y si no dar un mensaje que no fue encontrada

```python
def translate(word):
  return dic.get(word, "no encontrada")
```

Escribimos la palabra que queremos traducir y despues la imprimimos

```python
word = "si"
print(translate(word))
```
